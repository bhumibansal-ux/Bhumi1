```python
df = pd.read_csv('StudentPlacementData.csv')
df.shape
```

```python
df= df[df['Salary_LPA'] !=0]
```

```python
plt.figure(figsize=(15,5))
plt.scatter(df['CGPA'],df['Salary_LPA'], alpha=0.5, c= 'pink')
plt.xlabel('CGPA')
plt.ylabel('Salary_LPA)')
```

```python
X = df.iloc[:,3]
y = df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train.values.reshape(-1, 1), y_train)
```

```python
X_test
```

```python
y_test
```

```python
lr.predict(X_test.iloc[[0]].values.reshape(1,1))
```

```python
plt.figure(figsize=(15,5))
plt.scatter(df['CGPA'],df['Salary_LPA'], alpha=0.5, c= 'pink')
plt.plot(X_train,lr.predict(X_train.values.reshape(-1, 1)),color='maroon')
plt.xlabel('CGPA')
plt.ylabel('Salary_LPA)')
```

```python
m = lr.coef_
m
```

```python
b = lr.intercept_
b
```
